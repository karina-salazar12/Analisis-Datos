<h3 style= "color:#a7c5bd"> Problema de Negocio</h3>
<p> 
Actualmente, la institución enfrenta los siguientes desafíos:

Aumento en la morosidad: La tasa de incumplimiento de pago ha aumentado en los últimos años, impactando la rentabilidad del banco.
Decisiones basadas en criterios rígidos: El análisis crediticio se basa en reglas fijas que no consideran la variabilidad de cada cliente.
Demoras en el proceso de evaluación: La evaluación manual de cada solicitud toma tiempo y genera retrasos en la asignación de créditos.
La alta dirección del banco busca una solución basada en ciencia de datos para mejorar la toma de decisiones y reducir el riesgo crediticio.

Objetivo del Proyecto de Ciencia de Datos

El objetivo de este proyecto es desarrollar un modelo de Machine Learning que pueda predecir si un cliente incumplirá su pago el siguiente mes.

Las preguntas clave que el modelo debe responder son:

✅ ¿Cuáles son los factores más determinantes en la probabilidad de incumplimiento?</br>
✅ ¿Se puede predecir si un cliente incumplirá su pago el próximo mes?</br>
✅ ¿Cómo mejorar la precisión y eficiencia del proceso de evaluación crediticia?</br>

Para ello, se analizará un conjunto de datos que contiene información histórica de clientes, incluyendo su comportamiento de pago, límites de crédito, estado financiero y características personales.</p>
