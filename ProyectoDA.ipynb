```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

```python
dfventas=pd.read_csv("DataSet_ventas.csv",sep =";",low_memory=False)
dfventas.head(5)
```

```python
print(dfventas.columns)
```

```python
# Filtrar las columnas que contienen "Unnamed" en su nombre
columns_to_drop = [col for col in dfventas.columns if 'Unnamed' in col or ',' in col]

# Eliminar las columnas filtradas
dfventas.drop(columns=columns_to_drop, inplace=True)

# Reiniciar el Ã­ndice del dataframe
dfventas.reset_index(drop=True, inplace=True)

# Mostrar las primeras filas del dataframe modificado
print(dfventas.head())
```

```python
# Imprimir las columnas del DataFrame
print("Columnas antes de eliminar:", dfventas.columns)

# Lista de columnas innecesarias
columns_to_drop = ['Non Use Code', 'Assessor Remarks', 'OPM remarks', 'Location']  # Reemplaza con los nombres de las columnas que deseas eliminar

# Eliminar las columnas innecesarias
dfventas.drop(columns=columns_to_drop, inplace=True)

# Reiniciar el Ã­ndice del DataFrame
dfventas.reset_index(drop=True, inplace=True)

# Mostrar las primeras filas del DataFrame modificado
print("Columnas despuÃ©s de eliminar:", dfventas.columns)
print(dfventas.head())
```
