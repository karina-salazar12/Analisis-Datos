```python

# AsegÃºrate de que sale_amount es numÃ©rico
dfventas['sale_amount'] = pd.to_numeric(dfventas['sale_amount'], errors='coerce')

# AsegÃºrate de que data_recorded es tipo datetime
dfventas['data_recorded'] = pd.to_datetime(dfventas['data_recorded'], errors='coerce')

# Extraer el aÃ±o
dfventas['Year'] = dfventas['data_recorded'].dt.year

# Agrupar por aÃ±o y sumar correctamente
ingreso_anual = dfventas.groupby('Year')['sale_amount'].sum().reset_index(name='ingreso_total')

# Mostrar resultados
print(ingreso_anual)
```

```python
def agrupaYcuenta(df, columna, ordenar):
    return df.groupby(columna).size().reset_index(name='cantidad').sort_values(by='cantidad', ascending=ordenar)
```

```python
def agrupaYcuenta(dfParam, nomColumna, orden):
    gf = dfParam.groupby([nomColumna]).count()
    gf = gf.iloc[:, 0:1]  # Selecciona solo la primera columna
    gf.columns = ['cantidad']  # Renombra esa columna sin importar su nombre original
    gf = gf.sort_values("cantidad", ascending=orden)
    return gf
```

<h5>Cual es el costo promedio de las propiedades comerciales y residenciales?</h5>

```python
# Filtrar datos y calcular promedio
residential_avg = dfventas[dfventas['property_type'] == 'RESIDENTIAL']['assesed_value'].mean()
commercial_avg = dfventas[dfventas['property_type'] == 'COMMERCIAL']['assesed_value'].mean()

# Ajustar valores mÃ­nimos para evitar barras invisibles
commercial_avg = commercial_avg if commercial_avg > 0 else 1

# Crear DataFrame con los valores a graficar
data = {'Tipo de Propiedad': ['RESIDENTIAL', 'COMMERCIAL'],
        'Costo Promedio': [residential_avg, commercial_avg]}

# Graficar
fig, ax = plt.subplots(figsize=(8, 5), dpi=100)
ax.bar(data['Tipo de Propiedad'], data['Costo Promedio'], color=['#a2caa5', '#f4a460'])

# ConfiguraciÃ³n del eje Y para que siempre se vean las barras
ax.set_ylim(0, max(residential_avg, commercial_avg) * 1.1)

# ConfiguraciÃ³n del grÃ¡fico
ax.set_xlabel("Tipo de Propiedad", fontsize=12, fontweight="bold")
ax.set_ylabel("Costo Promedio", fontsize=12, fontweight="bold")
ax.set_title("GrÃ¡fico 4: Costo Promedio entre Propiedades Residenciales y Comerciales", fontsize=14, fontweight="bold")

# AÃ±adir texto en las barras
for i, valor in enumerate([residential_avg, commercial_avg]):
    ax.text(i, valor + 1, f'{valor:.2f}', ha='center', va='bottom', fontsize=10)

# Mostrar grÃ¡fico
plt.show()
```
